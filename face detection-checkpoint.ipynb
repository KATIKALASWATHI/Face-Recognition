# Face Recognition


**Here I used some students photos and and their names as input and detected the faces when WebCam is on.**
* *The detected faces names are saving to csv file.* *

```python
import cv2
import face_recognition
import numpy as np
import os
import pandas as pd
from os import listdir
os.getcwd()

os.chdir('C:\\Users\\Swathi\\Desktop\\photo')

cp=cv2.VideoCapture(0)
names=[]
image_names=[]
for filename in listdir('ph'):
    c=filename
    image_names.append(c)

fe1=[]
for image_name in image_names:
    li=face_recognition.load_image_file('ph/' + image_name)
    fe=face_recognition.face_encodings(li)[0]
    fe1.append(fe)


known_face_encodings=fe1
known_face_names=['kalpana',
 'Kiran',
 'nandini',
 'Neha',
 'Ram',
 'Roshan',
 'swathi']


while True:
    ret, frame=cp.read()
    rgb_frame=frame[:,:,::-1]
    
    face_locations=face_recognition.face_locations(rgb_frame)
    face_encodings=face_recognition.face_encodings(rgb_frame, face_locations)
    
    for(top, right, bottom, left), face_encoding in zip(face_locations, face_encodings):
        matches=face_recognition.compare_faces(known_face_encodings, face_encoding)
        name='Unknown'
        face_distances=face_recognition.face_distance(known_face_encodings, face_encoding)
        
        best_match_index=np.argmin(face_distances)
        if matches[best_match_index]:
            name=known_face_names[best_match_index]
        df=pd.DataFrame(names)
        df.to_csv('names.csv')
            
        
        cv2.rectangle(frame, (left,top),(right, bottom), (0,0,255,2))
        cv2.rectangle(frame, (left, bottom -35), (right, bottom), (0,0,255), cv2.FILLED)
        font=cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(frame, name, (left + 6, bottom-6), font, 1.0,(255,255,255),1)
        names.append(name)
        
    cv2.imshow('webcam_facerecognition', frame)
    

    if cv2.waitKey(10)==27:
        break
    
   
cp.release()
cv2.destroyAllWindows()




```
